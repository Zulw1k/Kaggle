```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
data = [[1, 'Buy', 'Add', 20.0, 100],
[2,'Sell','Add',25.0,200],
[3,'Buy','Add',23.0, 50],
[4,'Buy','Add',23.0, 70],
[3,'Buy','Remove',23.0, 50],
[5,'Sell' ,'Add', 28, 100],]
```

```python
trades = pd.DataFrame(data=data ,columns=['Id', 'Order', 'Type', 'Price', 'Quantity'])
```

```python
trades
```

```python
def display_sum(data):
    sum_of_buysell_orders = []
    best_price = []
    orders_count_buy = 0
    orders_count_sell = 0
    best_price_buy = float('inf')
    best_price_sell = float('-inf')
    for index, row in data.iterrows():
        if row['Order'] == 'Buy':
            if row['Price'] == best_price_buy:
                if row['Type'] == 'Add':
                    orders_count_buy +=1
                else:
                    orders_count_buy -=1
            elif row['Price'] > best_price_buy:
                orders_count_buy = 1
            else: 
                best_price_buy = row['Price']
                orders_count_buy = 1
            sum_of_buysell_orders.append(orders_count_buy)
            best_price.append(best_price_buy)
        else :
            if row['Price'] == best_price_sell:
                if row['Type'] == 'Add':
                    orders_count_sell +=1
                else:
                    orders_count_sell -=1
            elif row['Price'] < best_price_sell:
                orders_count_sell = 1
            else: 
                best_price_sell = row['Price']
                orders_count_sell = 1
            sum_of_buysell_orders.append(orders_count_sell)
            best_price.append(best_price_sell)
    return sum_of_buysell_orders, best_price
```

```python
def total_sum(data):
    total_sum_buy = []
    total_sum_sell = []
    total_sum = []
    row_sum = []
    the_sum = 0
    for index, row in data.iterrows():        
        the_sum = row['Price'] * row['Quantity'] 
        if row['Type'] == 'Add':
            row_sum.append(the_sum)
            total_sum.append(sum(row_sum))
            if row['Order'] == 'Buy':
                total_sum_buy.append(the_sum)
                total_sum_sell.append(0)
            else:
                total_sum_sell.append(the_sum)
                total_sum_buy.append(0)
        else:
            row_sum.append(-the_sum)
            total_sum.append(sum(row_sum))
            if row['Order'] == 'Buy':
                total_sum_buy.append(-the_sum)
                total_sum_sell.append(0)
            else:
                total_sum_sell.append(-the_sum)
                total_sum_buy.append(0)
    return total_sum , row_sum, total_sum_buy, total_sum_sell
            
```

```python
sum_of_buysell_orders, best_price = display_sum(trades)
```

```python
total_sum , row_sum, total_sum_buy, total_sum_sell = total_sum(trades)
```

```python
summary = pd.concat([trades,
           pd.DataFrame(sum_of_buysell_orders),
           pd.DataFrame(best_price),
           pd.DataFrame(total_sum),
           pd.DataFrame(row_sum),
           pd.DataFrame(total_sum_buy),
           pd.DataFrame(total_sum_sell)],
           axis = 1)
```

```python
summary.columns = ['Id', 'Order', 'Type', 'Price',
                  'Quantity', 'Number of orders', 'Best price',
                  'Total value', 'Order value', 'Total buy value',
                  'Total sell value']
```
